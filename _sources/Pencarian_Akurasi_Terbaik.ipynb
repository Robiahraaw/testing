```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pickle
```

```python
# Membaca data dari file CSV
data = pd.read_csv('glioma.csv')
```

```python
# Memisahkan fitur-fitur dan target
X = data[['Gender', 'Age_at_diagnosis', 'Race', 'IDH1', 'TP53', 'ATRX', 'PTEN', 'EGFR', 'CIC', 'MUC16', 'PIK3CA', 'NF1', 'PIK3R1', 'FUBP1', 'RB1', 'NOTCH1', 'BCOR', 'CSMD3', 'SMARCA4', 'GRIN2A', 'IDH2', 'FAT4', 'PDGFRA']]
y = data['Grade']
```

```python
# Normalisasi data menggunakan StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Simpan Standard Scaler ke dalam file pickle
with open('scaler_model.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)
```

```python
# Reduksi dimensi menggunakan PCA
pca = PCA(n_components=9)  # Anda bisa mengubah jumlah komponen sesuai kebutuhan
X = pca.fit_transform(X)
```

```python
# Simpan PCA model ke dalam file pickle
with open('pca_model.pkl', 'wb') as pca_file:
    pickle.dump(pca, pca_file)
```

```python
# Memisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Membuat model k-NN
k = 11  # Jumlah tetangga yang akan digunakan, Anda dapat mengubah nilai ini sesuai preferensi
knn_classifier = KNeighborsClassifier(n_neighbors=k)
```

```python
# Melatih model
knn_classifier.fit(X_train, y_train)
```

```python
# Melakukan prediksi
y_pred = knn_classifier.predict(X_test)
```

```python
# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)
```

```python
# Simpan model K-NN ke dalam file pickle
with open('knn_model.pkl', 'wb') as knn_file:
    pickle.dump(knn_classifier, knn_file)
```
