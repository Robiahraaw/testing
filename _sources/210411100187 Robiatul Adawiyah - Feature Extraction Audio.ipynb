```python
from scipy.stats import skew, norm, gamma
np.random.seed(42)

# generate normal, left-skewed, and right-skewed samples
normal_freqs = np.random.normal(loc=0, scale=1, size=1000)
left_skewed_freqs = np.random.weibull(50,1000)
right_skewed_freqs = gamma.rvs(a=1, size=1000)

# plot these
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(14, 4))
axs[0].hist(left_skewed_freqs, bins=30, density=True, alpha=0.7)
axs[0].set_title("Left Skewness")
axs[1].hist(left_skewed_freqs, bins=30, density=True, alpha=0.7)
axs[1].set_title("Normal Skewness")
axs[2].hist(right_skewed_freqs, bins=30, density=True, alpha=0.7)
axs[2].set_title("Right Skewness")
plt.tight_layout()
plt.show()
```
