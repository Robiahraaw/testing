Klasifikasi Emosi Pada Audio

```python
import pandas as pd
import numpy as np
import librosa
from scipy.stats import kurtosis, skew
```

```python
# Daftar nama file audio
audio_files = ['a01.wav', 'a02.wav', 'a03.wav', 'a04.wav', 'a05.wav', 'd01.wav', 'd02.wav', 'd03.wav', 'd04.wav', 'd05.wav', 'f01.wav', 'f02.wav', 'f03.wav', 'f04.wav', 'f05.wav', 'h01.wav', 'h02.wav', 'h03.wav', 'h04.wav', 'h05.wav', 'n01.wav', 'n02.wav', 'n03.wav', 'n04.wav', 'n05.wav', 'sa01.wav', 'sa02.wav', 'sa03.wav', 'sa04.wav', 'sa05.wav', 'su01.wav', 'su02.wav', 'su03.wav', 'su04.wav', 'su05.wav',]
```

```python
# Daftar Label
label = ['angry', 'angry', 'angry', 'angry', 'angry', 'disgust', 'disgust', 'disgust', 'disgust', 'disgust', 'fear', 'fear', 'fear', 'fear', 'fear', 'happy', 'happy', 'happy', 'happy', 'happy', 'neutral', 'neutral', 'neutral', 'neutral', 'neutral', 'sad', 'sad', 'sad', 'sad', 'sad', 'surprise', 'surprise', 'surprise', 'surprise', 'surprise']
```

```python
# Inisialisasi list untuk menyimpan statistik ZCR
zcr_mean_values = []
zcr_median_values = []
zcr_std_values = []
kurtosis_values = []
skewness_values = []
```

```python
# Loop melalui setiap file audio
for audio_file in audio_files:
    # Membaca file audio
    y, sr = librosa.load(audio_file, sr=None)

    # Menghitung ZCR
    zcr = librosa.feature.zero_crossing_rate(y)

    # Menghitung statistik ZCR
    zcr_mean = np.mean(zcr)
    zcr_median = np.median(zcr)
    zcr_std = np.std(zcr)

    # Menghitung kurtosis dan skewness
    kurt = kurtosis(zcr.ravel())
    skewness = skew(zcr.ravel())

    # Menambahkan statistik ZCR ke dalam list
    zcr_mean_values.append(zcr_mean)
    zcr_median_values.append(zcr_median)
    zcr_std_values.append(zcr_std)
    kurtosis_values.append(kurt)
    skewness_values.append(skewness)
```

```python
# Membuat DataFrame dengan fitur audio dan statistik ZCR
data = {'Audio File': audio_files, 'ZCR Mean': zcr_mean_values, 'ZCR Median': zcr_median_values, 'ZCR Std': zcr_std_values, 'Kurtosis': kurtosis_values, 'Skewness': skewness_values, 'Labels': label}
df = pd.DataFrame(data)
```
